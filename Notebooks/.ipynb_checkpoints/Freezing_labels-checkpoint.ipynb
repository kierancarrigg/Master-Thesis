```python
import os
import pandas as pd
import numpy as np
import re
```

```python
# Function to extract animal ID from file name
def extract_animal_id(file_name):
    match = re.search(r'Animal (\d+)_(\d+)\.csv', file_name)
    if match:
        return '.'.join(match.groups())
    else:
        match = re.search(r'Animal_(\d+)_(\d+)\.csv', file_name)
        if match:
            return '.'.join(match.groups())
        else:
            return None

# Directory containing CSV files
directory = "../Datasets/Freezing_predictions"

# Initialize an empty dictionary to store data
data_dict = {}

# Loop through files in the directory
for file_name in os.listdir(directory):
    if file_name.endswith(".csv"):
        # Construct the full path to the file
        file_path = os.path.join(directory, file_name)

        # Extract animal ID
        animal_id = extract_animal_id(file_name)

        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)

        # Extract the last column
        last_column = df.iloc[:, -1]

        # Add the data to the dictionary with animal ID as the key
        if animal_id is not None:
            data_dict[animal_id] = last_column

# Example usage: Access data using animal ID
if "11.4" in data_dict:
    print(f"Animal ID: 11.4, Last Column Data:\n{data_dict['11.4']}")
else:
    print("Animal ID 11.4 not found in the data dictionary.")
```

```python
# Path to the folder containing the data files
folder_path = '../Datasets/MouseDirectionalPolar'

# Iterate through files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)

        # Extract mouse ID from the file name (assuming the format is "X.Y_Original_data.csv")
        mouse_id = filename.split('_')[0]

        # Check if the mouse ID exists in your dictionary
        if mouse_id in data_dict:
            # Load the data file
            data = pd.read_csv(file_path)

            # Extract the corresponding mouse data from the dictionary
            mouse_data = data_dict[mouse_id]

            # Add a new column to the data file with the loaded mouse data
            data['Freezing_label'] = mouse_data  # Replace 'YourExistingColumn' with the appropriate column name

            # Save the modified data file
            data.to_csv(file_path, index=False)
```
